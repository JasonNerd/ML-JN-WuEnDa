# 从0开始的深度卷积神经网络搭建--AlexNet
1. 如下图, LeNet包含两个卷积块，每一个卷积块都是“卷积-激活-池化”的三层结构, 通道数从6增加到16，卷积核大小均为5，激活函数使用sigmoid.最后展平并使用几个全连接层将输出降到10维。


2. AlexNet开启了深度卷积网络的研究热潮(2012)，原因之一是**ImageNet发布了百万量级的标注数据集**，其二是**可编程图形处理器**的应用。注意到ImageNet数据集的图片都是BGR3通道的，并且宽高都更大，例如224x224


```python
import torch
from torch import nn
import sys
sys.path.append("../")
from MyUtils import d2laiUtils as d2l
```

```python
# 1. 依据图示构造AlexNet
alex_net = nn.Sequential(
    nn.Conv2d(1, 96, kernel_size=11, stride=4, padding=1), nn.ReLU(), # 54x54
    nn.MaxPool2d(kernel_size=3, stride=2),  # 96x26x26
    nn.Conv2d(96, 256, kernel_size=5, padding=2), nn.ReLU(),  # 256x26x26
    nn.MaxPool2d(kernel_size=3, stride=2), # 256x12x12
    nn.Conv2d(256, 384, kernel_size=3, padding=1), nn.ReLU(), # 384x12x12
    nn.Conv2d(384, 384, kernel_size=3, padding=1), nn.ReLU(), # 284x12x12
    nn.Conv2d(384, 256, kernel_size=3, padding=1), nn.ReLU(), # 256x12x12
    nn.MaxPool2d(kernel_size=3, stride=2), # 256x5x5 = 6400
    nn.Flatten(),
    nn.Linear(6400, 4096), nn.ReLU(),
    nn.Dropout(0.5),
    nn.Linear(4096, 4096), nn.ReLU(),
    nn.Dropout(0.5),
    nn.Linear(4096, 10)
)
alex_net
```

```python
X = torch.randn(1, 1, 224, 224)
for layer in alex_net:
    X = layer(X)
    print(f"{layer.__class__.__name__} out-shape={X.shape}")
```

```python
# 训练AlexNet
batch_size = 256
train_iter, test_iter = d2l.loadFashionMnist(batch_size, resize=224)
lr, num_epochs, gpu, cpu = 0.01, 10, torch.device("cuda:0"), torch.device("cpu")
d2l.train_ch6(alex_net, train_iter, test_iter, num_epochs, lr, gpu)
```
