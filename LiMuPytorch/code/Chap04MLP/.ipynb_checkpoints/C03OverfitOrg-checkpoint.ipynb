```python
import math
import numpy as np
import torch
from torch import nn
import sys
sys.path.append("../")
from MyUtils import d2laiUtils as d2l
from torch.utils import data

def polyDatasetGen(n, w_true, batch_size, is_train=True):
    """依据w系数向量(d)生成大小为nxd的输入以及nx1的输出, 返回dataset对象"""
    x1 = np.random.normal(size=(n, 1))
    X = np.power(x1, range(len(w_true)))
    for i in range(len(w_true)):
        X[:, i] /= math.gamma(i+1) # scaling
    y = np.dot(X, w_true)
    y += np.random.normal(scale=0.5, size=y.shape)
    X = torch.tensor(X, dtype=torch.float32)
    y = torch.tensor(y, dtype=torch.float32)
    print(X.shape, y.shape)
    dataset = data.TensorDataset(X, y)
    return data.DataLoader(dataset, shuffle=is_train, batch_size=batch_size)

def lossEvalute(data_iter, loss, model):
    """返回模型在指定数据集上的平均损失"""
    cnt = d2l.Accumulator(2)
    for X,y in data_iter:
        out = model(X)
        y = y.reshape(out.shape)
        l = loss(out, y)
        cnt.add(l.sum(), l.numel())
    return cnt[0] / cnt[1]

def train(train_iter, test_iter, num_epoch=400):
    """在指定数据集上训练, 返回训练好的model以及在test_iter上的损失情况"""
    size_dt = t_d.dataset.tensors[0].shape[0]
    num_inputs = t_d.dataset.tensors[0].shape[1]
    model = nn.Sequential(nn.Linear(num_inputs, 1))
    trainer = torch.optim.SGD(model.parameters(), lr=0.01)
    loss = nn.MSELoss(reduction="None")
    train_loss, test_loss = [], []
    for epoch in range(num_epoch):
        for X, y in train_iter:
            out = model(X)
            y = y.reshape(out)
            l = loss(out, y)
            trainer.zero_grad()
            l.sum().backward()
            trainer.step()
        with torch.no_grad():
            train_loss.append(lossEvalute(train_iter, loss, model))
            test_loss.append(lossEvalute(test_iter, loss, model))
    return model, train_loss, test_loss

def demostrate():
    powerNum = 20
    train_size, test_size = 200, 800
    batch_size = min(10, train_size)
    w_true = np.zeros(powerNum)
    w_true[:4] = np.array([12.09, 2.26, 18.06, 2.018])
    train_iter = polyDatasetGen(train_size, w_true, batch_size)
```
