# 深度循环神经网络
对于每一个输出的隐状态，进行一个多连接


```python
import torch
from torch import nn
import sys
sys.path.append("../")
from MyUtils import d2laiUtils as d2l
```

```python
batch_size, num_steps, use_random = 128, 32, False
train_iter, vocab = d2l.load_data_time_machine(batch_size, num_steps, use_random, -1)
vocab_size, num_hiddens, num_layers = len(vocab), 256, 3
lstm_layer = nn.LSTM(vocab_size, num_hiddens)
device = torch.device("cuda:0")
model = d2l.RNNModel(lstm_layer, vocab_size)
model.to(device)
```

```python
lr, num_epochs = 2, 500
d2l.train_ch8(model, train_iter, vocab, lr, num_epochs, device, use_random)
```

```python
d2l.predict_ch8("feel good", 100, model, vocab, device)
```
