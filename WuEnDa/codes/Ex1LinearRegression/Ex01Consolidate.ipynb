```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# 1. 查看原始数据的情况, header names
data = pd.read_csv('./ex1data1.txt', header=None, names=['population', 'profit'])
# data.head() # 前几行
# data.info() # 97x2(float)
data.plot(kind='scatter', c='red', marker='x', x='population', y='profit')
```

```python
# # 2. 构造矩阵X, y, theta
# index, name, value of a column[warning: you should only run it onece]
data.insert(0, 'ones', 1) 
```

```python
n = data.shape[1]
X = np.matrix(data.iloc[:,0:n-1]) # 所有的行, 取前n-1列
y = np.matrix(data.iloc[:,n-1:n]) # 所有的行, 取最后一列
theta = np.matrix(np.zeros(n-1)) # 初值为0, 0
# X.head(), y.head() # 查看划分是否正确
X.shape, y.shape, theta.shape
```

$h(X) = X*\theta^T, J_{(\theta_0, \theta_1)} = \frac{innnerSum([h(X)-y]^2)}{2*m}$

```python
# 3. 构造costFunc(X, y, theta)
def costFunc(X, y, theta):
    m = len(y)
    err = X*theta.T-y
    return np.sum(np.power(err, 2))/(2*m)
costFunc(X,y,theta) # 看一下初值
```

```python
# 4. 初始化 迭代次数, 学习速率alpha, 设计batch gradient decent
iter_n = 300
alpha = 0.018
def gradientDecent(X, y, theta, alpha, iter_n):
    m = len(y)
    cost = np.zeros(iter_n)
    for i in range(iter_n):
        err = X*theta.T-y
        delta = (X.T*err).T/m
        theta = theta - alpha*delta
        cost[i] = costFunc(X, y, theta)
    return theta, cost
```

```python
#  5. 设计batch gradient decent, 得出结果, 作出拟合曲线, 作出 代价-迭代次数曲线
theta, cost = gradientDecent(X, y, theta, alpha, iter_n)
theta
```

```python
cost[-1]
```

```python
fig, ax = plt.subplots() # 绘制多种曲线
ax.scatter(data.population, data.profit, c='red', marker='x') # 注意操作 .
x = np.linspace(data.population.min(), data.population.max(), 100) # 注意操作 linspace
h = theta[0,0]+theta[0,1]*x
ax.plot(x, h)
plt.show()
```

```python
plt.plot(range(iter_n), cost) 
# 可以截取cost的不同部分查看, 例如前10次, 实际上在3到4次的时候就变得很小了（相对）
# 此后减小的速度大大放缓
plt.show()
```

实际上以上定义的函数对于多变量回归分析也是适用的, 例如还有一个数据集，其中有2个变量（房子的大小，卧室的数量）和目标（房子的价格）但此时就难以可视化了只能观察代价曲线

```python
data2 = pd.read_csv('./ex1data2.txt', header=None, names=['Area', 'Bedrooms', 'Price'])
data2 = (data2-data2.mean())/data2.std() # 特征归一化(feature scaling)
data2.insert(0, 'Ones', 1)
data2.head()
```

```python
n2 = data2.shape[1]
X2 = np.matrix(data2.iloc[:, 0:n2-1])
y2 = np.matrix(data2.iloc[:, n2-1:n2])
theta2 = np.matrix(np.zeros(n2-1))
X2.shape, y2.shape, theta2.shape
```

```python
g2, cost2 = gradientDecent(X2, y2, theta2, 0.02, 500)
g2
```
