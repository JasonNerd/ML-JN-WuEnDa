```python
# pandas读取数据集, 查看数据基本信息
flstr = 'ex1data1.txt'
data = pd.read_csv('ex1data1.txt', header=None, names=['Population', 'Profits']) 
data.head()
```

**接下来是实现代价函数**(注意使用向量机制实现)
$$J\left( \theta  \right)=\frac{1}{2m}\sum\limits_{i=1}^{m}{{{\left( {{h}_{\theta }}\left( {{x}^{(i)}} \right)-{{y}^{(i)}} \right)}^{2}}}$$
其中：\\[{{h}_{\theta }}\left( x \right)={{\theta }^{T}}X={{\theta }_{0}}{{x}_{0}}+{{\theta }_{1}}{{x}_{1}}+{{\theta }_{2}}{{x}_{2}}+...+{{\theta }_{n}}{{x}_{n}}\\] 

```python
# 对于单变量而言 h(xi) = t0 + t1*xi = (1, xi) x (t0, t1)^T
# x = [x1 ... xm].T(mx1), X是一个m行1列的列向量
# 添加一列 ones = [1 ... 1].T(mx1), X = [ones, x](mx2)
# y = [y1 ... ym].T(mx1)
# theta = [t0, t1](1x2) so that (h(x)-y) = X*theta.T - y
# 注意它们都应该是 np.matrix 类型, 而不是 np.array 类型
def costFunc(X, y, theta):
    inner = np.power(X*theta.T-y, 2)
    return np.sum(inner)/(2*len(X))
# here we use len() function to get m = len(X)
```

```python
# 在原始数据中添加一列1
# data.insert(0, 1, 'ones') (column index, column name, column content)
# data.drop(1, axis=1, inplace=True) # we del the wrong insertion 
data.insert(0, 'ones', 1)
data.head()
```

```python
# 划分X和y
col_n = data.shape[1]
X = data.iloc[:, 0:col_n-1];
y = data.iloc[:, col_n-1:col_n]
X.head(), y.head() # 查看划分的正确性
```

```python
# 转为matrix以进行计算
X = np.matrix(np.array(X))
y = np.matrix(np.array(y))
theta = np.matrix(np.array([0, 0]))
X.shape, y.shape, theta.shape # 查看矩阵形状
```

```python
costFunc(X, y, theta) # 初始代价值
```

## gradient descent
1. 更新公式: $\theta j(new) = \theta j(old) - \alpha*d(\theta j)$
2. 在本问题中, j = 0 or 1, $\alpha$为学习率(learning rate), d(x)表示对x的(偏)微分
3. 注意同步更新

```python
def gradientTheta(X, y, theta, alpha):
    # 对 theta 做一次更新, 计算costFunc并返回theta(new), loss
    m = len(X)
    n = X.shape[1]
    dtheta_0 = np.sum(X*theta.T-y)/m
    dtheta_1 = np.sum(X[:,n-1:n].T*(X*theta.T-y))/m
    dtheta = np.matrix([dtheta_0, dtheta_1])
    loss = costFunc(X, y, theta)
    return theta-alpha*dtheta, loss
```

```python
# 做多次迭代
def train(iter_n, X, y, theta, alpha):
    for i in range(iter_n):
        theta, loss = gradientTheta(X, y, theta, alpha)
        losses.append(loss)
        theta_0s.append(float(theta[0:1, 0:1]))
        theta_1s.append(float(theta[0:1, 1:2]))
```

```python
# 配置迭代次数, 学习率, 进行迭代计算
iter_n = 1000
alpha = 0.012
losses = []
theta_0s = []
theta_1s = []
train(iter_n, X, y, theta, alpha)
```

```python
losses[-1]
```

```python
# 作出拟合图像
x = np.linspace(data.Population.min(), data.Population.max(), 100)
theta_0 = theta_0s[-1]
theta_1 = theta_1s[-1]
hx = theta_0 + theta_1*x 
# type(x) # numpy.ndarray 支持浮点操作

# 画出拟合直线 和 原始数据散点
fig, ax = plt.subplots(figsize=(12,8))
ax.plot(x, hx, label='Prediction')
ax.scatter(data.Population, data.Profits, label='Traning Data', c='red', marker='x')
ax.legend(loc=2)
ax.set_xlabel('Population')
ax.set_ylabel('Profits')
ax.set_title('Predicted Profits vs. Population Size')
plt.show()
```

```python
# 绘制theta变化
fig, ax = plt.subplots(figsize=(12,8))
ax.plot(range(len(losses)), losses, label='loss value')
ax.legend(loc=1)
ax.set_xlabel('iteration times')
ax.set_ylabel('loss value')
ax.set_title('loss value changes with gradient descent')
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(12,8))
ax.plot(range(len(theta_0s)), theta_0s, label='theta_0')
ax.plot(range(len(theta_1s)), theta_1s, label='theta_1')
ax.legend(loc=2)
ax.set_xlabel('iteration times')
ax.set_ylabel('aarguments value')
ax.set_title('arguments value changes with gradient descent')
plt.show()
```
